## Baseline characteristics

Commonly, researchers assemble demographic characteristics of groups, broadly called "baseline characteristics".
There is a lot that falls under this umbrella term, so here is a fairly minimal version that gets the gender balance of the groups, the ethnicities, and a summary of the ages.

First, we will write a query that will get us the values we need.

```python
baseline_query = """
WITH condition_group AS (
  SELECT person_id,
      EXTRACT(days FROM '2025-09-17'::date - birth_datetime)::int as days_age,
      CASE WHEN gender_concept_id = 8532 THEN 1 ELSE null END as female,
      CASE WHEN gender_concept_id = 8507 THEN 1 ELSE null END as male,
      CASE WHEN race_concept_id = 8527 THEN 1 ELSE null END AS white,
      CASE WHEN race_concept_id = 8516 THEN 1 ELSE null END AS black_or_african_american,
      CASE
        WHEN person_id IN (
          SELECT person_id
          FROM "DelphiDemo".condition_occurrence
          WHERE condition_concept_id = 139750
        ) THEN 'with'
        ELSE 'without'
      END AS condition_status
  FROM "DelphiDemo".person
)

SELECT  condition_status,
        COUNT(person_id) as total,
        SUM(male) as male,
        SUM(female) as female,
        SUM(white) as white,
        SUM(black_or_african_american) as black_or_african_american,
        SUM(days_age) as total_age,
        SUM(days_age^2) as total_age_sq,
        MIN(days_age) as minimum_age,
        MAX(days_age) as maximum_age
FROM condition_group
GROUP BY condition_status;
"""
```

The general approach here is to make a column with `1` or `null` for different categorical variables (race_concept_id, gender_concept_id in this case) in the first Common Table Expression (CTE), then sum those columns to get the count for that category by condition.

To get the ages of people, we're getting the number of days from birth until the date of publication for the [paper accompanying the model](https://www.nature.com/articles/s41586-025-09529-3) used to generate the data.

We then use a TES message builder to submit this query.

```python
wb.build_tes.simple_sql(
    name="Baseline characteristics",
    query=baseline_query
)

#wb.submit()
```

Then we can fetch the outputs of this task, and load the data from the output files.

```python
#wb.fetch_outputs()
```

```python
# The paths will depend on the task ID of your submission
baseline_characteristics_paths = ["./output/Nottingham TRE 01/1056/output.csv", "./output/Nottingham TRE 02/1057/output.csv"]
```

The baseline characteristics can be assembled from the summary statistics from the query.

```python
def baseline_characteristics(paths: list[Path]):
    data = [pd.read_csv(path) for path in paths]

    # Use the condition status to get the groups
    data_with = [table[table["condition_status"] == "with"].iloc[0] for table in data]
    data_without = [table[table["condition_status"] == "without"].iloc[0] for table in data]

    # Collect the categorical variables
    categorical_with = (
        data_with[0][["total", "male", "female", "white", "black_or_african_american"]] +
        data_with[1][["total", "male", "female", "white", "black_or_african_american"]]
        )
    categorical_with["other"] = categorical_with["total"] - categorical_with["white"] - categorical_with["black_or_african_american"]

    categorical_without = (
        data_without[0][["total", "male", "female", "white", "black_or_african_american"]] +
        data_without[1][["total", "male", "female", "white", "black_or_african_american"]]
        )
    categorical_without["other"] = categorical_without["total"] - categorical_without["white"] - categorical_without["black_or_african_american"]

    # Calculate percentages
    proportions_with = categorical_with/categorical_with["total"] * 100
    proportions_without = categorical_without/categorical_without["total"] * 100

    # Collect the days for age statistics
    days_with = (
        data_with[0][["total", "total_age", "total_age_sq"]] +
        data_with[1][["total", "total_age", "total_age_sq"]]
    )

    days_without = (
        data_without[0][["total", "total_age", "total_age_sq"]] +
        data_without[1][["total", "total_age", "total_age_sq"]]
    )

    mean_age_with = days_with["total_age"]/(days_with["total"]*365)
    mean_age_without = days_without["total_age"]/(days_without["total"]*365)

    std_age_with = math.sqrt(
        days_with["total_age_sq"] / days_with["total"] -
        (days_with["total_age"] / days_with["total"]) ** 2
    )/365
    std_age_without = math.sqrt(
        days_without["total_age_sq"] / days_without["total"] -
        (days_without["total_age"] / days_without["total"]) ** 2
    )/365

    # These age statistics are not aggregated by adding together
    min_age_with = min(x["minimum_age"] for x in data_with)/365
    min_age_without = min(x["minimum_age"] for x in data_without)/365

    max_age_with = max(x["minimum_age"] for x in data_with)/365
    max_age_without = max(x["minimum_age"] for x in data_without)/365

    # Render as a nice-ish table
    return Markdown(
        f"""
### Baseline Characteristics

| | With | Without |
| :--- | :---|---|
| Total | {categorical_with["total"]} | {categorical_without["total"]} |
| Gender |
| Male (%) | {proportions_with["male"]:.2f}% | {proportions_without["male"]:.2f}% |
| Female (%) | {proportions_with["female"]:.2f}% | {proportions_without["female"]:.2f}% |
| Ethnicity |
| White | {proportions_with["white"]:.2f}% | {proportions_without["white"]:.2f}% |
| Black or African American | {proportions_with["black_or_african_american"]:.2f}% | {proportions_without["black_or_african_american"]:.2f}% |
| Other | {proportions_with["other"]:.2f}% | {proportions_without["other"]:.2f}% |
| Age |
| Mean (± std)| {mean_age_with:.2f} ({std_age_with:.2f}) | {mean_age_without:.2f} ({std_age_without:.2f}) |
| Minimum | {min_age_with:.2f} | {min_age_without:.2f}|
| Maximum | {max_age_with:.2f} | {max_age_without:.2f}|
        """
    )

baseline_characteristics(baseline_characteristics_paths)
```

## T-test

Now we can write a SQL query that will let us perform a t-test comparing the blood pressure between people with and without primary neoplasm of skin.
The only statistics we need for this are the count, the sum, and sum of squares of blood pressure for each group.

The query itself is in three parts.
The first, which defines the CTE `last_occurrence` makes sure that for each person we only take the last measurement.
The second, which defines the CTE `value_with_status` splits these measurements into measurements "with" and "without" primary neoplasm of skin.
The final part gathers these into a table that gets the count, sum, and sum of squares for these groups.

```python
sys_pressure_neoplasm_query = """
WITH last_occurrence AS (
    SELECT value_as_number,
           ROW_NUMBER() OVER (
               PARTITION BY person_id
               ORDER BY measurement_datetime DESC NULLS LAST
           ) AS rn,
           person_id
    FROM "DelphiDemo".measurement
    WHERE measurement_concept_id = 3004249
),
  
value_with_status AS (
  SELECT value_as_number,
       CASE
        WHEN person_id IN (
          SELECT person_id
          FROM "DelphiDemo".condition_occurrence
          WHERE condition_concept_id = 139750
        ) THEN 'with'
        ELSE 'without'
       END AS condition_status
  FROM last_occurrence
  WHERE rn = 1
)

SELECT  condition_status,
        COUNT(value_as_number) as count,
        SUM(value_as_number) as sum,
        SUM(value_as_number^2) as sumsq
FROM value_with_status
GROUP BY condition_status;
"""
```

Then we can use another TES message builder to submit this query.

```python
wb.build_tes.simple_sql(
    name="Systolic blood pressure comparison",
    query=sys_pressure_neoplasm_query
)

#wb.submit()
```

The `partialstats` features we imported at the top provide utilities for doing federated statistics.
It defines `Partial` results which can be combined to give the result for the whole set.
All we need to do is define a function to build a `SumOfSquaresPartial` and we have everything we need.

The result of the task we submitted earlier is a table like this:

```python
pd.read_csv("./output/Nottingham TRE 01/1039/output.csv")
```

The `collect_var_data` function builds a `SumOfSquaresPartial` for rows of a table like this.
If you have some other groupings in your data, you can choose other `group_var` and `group_level` variables.

```python
def collect_var_data(path: Path, group_var: str = "condition_status", group_level: str = "with") -> SumOfSquaresPartial:
    data = pd.read_csv(path)
    row_data = data[data[group_var] == group_level]
    return SumOfSquaresPartial(
        count=row_data["count"].iloc[0],
        sum=row_data["sum"].iloc[0],
        sumsq=row_data["sumsq"].iloc[0]
    )

collect_var_data("./output/Nottingham TRE 01/1039/output.csv", group_level="with")
```

The `SumOfSquaresPartial` can be used to get different summary statistics using the imported combiner functions.
Here, they are wrapped in a class that can be used to configure different groupings.

```python
class GroupStatsAggregator:
    def __init__(
        self,
        paths: list[Path],
        group_var: str = "condition_status",
        group_levels: list[str] = ["with", "without"]
    ):
        self._paths = paths
        self._group_var = group_var
        self._group_levels = group_levels
        self._data: dict[str, SumOfSquaresPartial] = self.collect_data()

    def collect_data(self):
        return {
            gl:[collect_var_data(p, group_var=self._group_var, group_level=gl) for p in self._paths]
            for gl in self._group_levels
        }

    def data(self, key: str) -> list[SumOfSquaresPartial]:
        return self._data[key]

    @property
    def count(self) -> dict[str, int]:
        return {
            gl: sum_combiner.run()
            for gl, value in self._data.items()
        }
        
    @property
    def mean(self) -> dict[str, np.float64]:
        return {
            gl: mean_combiner.combine(value)
            for gl, value in self._data.items()
        }

    @property
    def variance(self) -> dict[str, np.float64]:
        return {
            gl: variance_combiner.combine(value)
            for gl, value in self._data.items()
        }

    @property
    def std(self) -> dict[str, np.float64]:
        var = self.variance
        return {gl: np.sqrt(value) for gl, value in var.items()}
    
```

The data can then be loaded from the output files.

```python
data_paths = [Path("./output/Nottingham TRE 01/1039/output.csv"), Path("./output/Nottingham TRE 02/1040/output.csv")]
agg = GroupStatsAggregator(data_paths)
```

`scipy.stats` includes a function for calculating an independent t-test from the mean, standard deviation and count of two groups.
Here, the large sample size means we will probably get a significant p-value, so a calculation of Cohen's $d$ is included for effect size.

```python
@dataclass
class Sample:
    name: str
    data: list[SumOfSquaresPartial]

    @property
    def stats(self) -> tuple[int, float, float]:
        return (
            sum_combiner.combine(self.data),
            mean_combiner.combine(self.data),
            variance_combiner.combine(self.data)
        )

class TwoSampleComparator:
    def __init__(
        self,
        sample_1: Sample,
        sample_2: Sample
    ) -> None:
        self._sample_1 = sample_1
        self._sample_2 = sample_2

    def two_sample_t_stat(self, equal_var: bool = False) -> tuple[float, int]:
        n_a, mean_a, var_a = self._sample_1.stats
        n_b, mean_b, var_b = self._sample_2.stats
        
        std_a = np.sqrt(var_a)
        std_b = np.sqrt(var_b)
        return stats.ttest_ind_from_stats(
            mean_a, std_a, n_a,
            mean_b, std_b, n_b,
            equal_var=equal_var
        )

    def cohen_d(self) -> np.float64:
        n_1, x_1, s_1 = self._sample_1.stats
        n_2, x_2, s_2 = self._sample_2.stats
        s = math.sqrt(((n_1 - 1) * s_1 + (n_2 - 1) * s_2) / (n_1 + n_2 - 2))
        return abs(x_1 - x_2) / s

    def report(self, equal_var: bool = False) -> Markdown:
        t_test = self.two_sample_t_stat(equal_var)
        d = self.cohen_d()
        stats_1 = self._sample_1.stats
        stats_2 = self._sample_2.stats
        return Markdown(
            f"""
|   | {self._sample_1.name} | {self._sample_2.name} |
| --- | --- | --- |
| count | {stats_1[0]} | {stats_2[0]} |
| mean | {stats_1[1]:.2f} | {stats_2[1]:.2f} |
| std | {np.sqrt(stats_1[2]):.2f} | {np.sqrt(stats_2[2]):.2f} |

$P$ = {t_test.pvalue:.2f}

Cohen's $d$ = {d:.2f}
            """
        )
```

```python
compare = TwoSampleComparator(
    Sample(name="Patients with primary neoplasm of skin", data=agg.data("with")),
    Sample(name="Patients without primary neoplasm of skin", data=agg.data("without"))
)

compare.report()
```

If this were real data, we could conclude, with a $P$ > 0.001 and a Cohen's $d$ of 0.58, that we have found a real result!

## Contingency table

So patients with primary neoplasm of skin have a slightly higher systolic blood pressure than those without.
Let's see if there's an association between primary neoplasm of skin and hypertension.

```python
codesets.plot_by_codes(codesets.get_codes_by_substring_match("hypertension")["OMOP"])
```

Now we have the concept_id for hypertension (320128), we can write a query to get counts of people with/without hypertension and primary neoplasm of skin and do a Fisher's exact test on the resulting table.

```python
hypertension_query = """
WITH hypertension AS (
  SELECT
    person_id,
    CASE
      WHEN person_id IN (
        SELECT person_id
        FROM "DelphiDemo".condition_occurrence
        WHERE condition_concept_id = 320128
      ) THEN 'has_hypertension' ELSE 'no_hypertension' END AS hypertension_status
  FROM "DelphiDemo".person
)

SELECT
  CASE
    WHEN p.person_id IN (
      SELECT person_id
      FROM "DelphiDemo".condition_occurrence
      WHERE condition_concept_id = 139750
    ) THEN 'with'
    ELSE 'without'
    END AS neoplasm_status,
  hypertension.hypertension_status,
  COUNT(p.person_id)
FROM "DelphiDemo".person p
  JOIN hypertension ON p.person_id = hypertension.person_id
GROUP BY neoplasm_status, hypertension_status
"""

wb.build_tes.simple_sql(
    name="hypertension x neoplasm",
    query=hypertension_query
)

#wb.submit()
```
